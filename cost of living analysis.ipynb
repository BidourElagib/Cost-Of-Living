Importing the libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

Loading the dataset:

```python
df = pd.read_csv('cost-of-living.csv')
```

Checking for and handling any missing data:

```python
df.describe().round(2)
```

```python
print(df.isna().sum())
```

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
print(df.isna().sum())
```

Handling duplicate data:

```python
print(df.duplicated().sum())
df_cleaned = df.drop_duplicates()
```

Feature engineering:

```python
df_cleaned['dining_out_cost'] = (
    df_cleaned['x1'] + df_cleaned['x2'] + df_cleaned['x3'] + df_cleaned['x4'] + df_cleaned['x5'] + df_cleaned['x6'] + df_cleaned['x7'] +df_cleaned['x8']
)
df_cleaned['grocery_cost'] = (
    df_cleaned['x9'] + df_cleaned['x10'] + df_cleaned['x11'] + df_cleaned['x12'] + df_cleaned['x13'] + df_cleaned['x14'] + df_cleaned['x15'] +
    df_cleaned['x16'] + df_cleaned['x17'] + df_cleaned['x18'] + df_cleaned['x19'] + df_cleaned['x20'] + df_cleaned['x21'] + df_cleaned['x22'] + 
    df_cleaned['x23'] + df_cleaned['x24'] + df_cleaned['x25'] + df_cleaned['x26'] + df_cleaned['x27']
)
df_cleaned['avg_transport_cost'] = (
    df_cleaned['x28'] + df_cleaned['x29'] + df_cleaned['x30'] + df_cleaned['x31'] + df_cleaned['x32'] + df_cleaned['x33'] + df_cleaned['x34'] + df_cleaned['x35']
) / 8

df_cleaned['utility_cost'] = (
    df_cleaned['x36'] + df_cleaned['x37'] + df_cleaned['x38']
)

df_cleaned['avg_entertainment_cost'] = (
    df_cleaned['x39'] + df_cleaned['x40'] + df_cleaned['x41']
) / 3

df_cleaned['education_cost'] = (
     df_cleaned['x42'] + df_cleaned['x43'] 
)

df_cleaned['clothing_cost'] = (
    df_cleaned['x44'] + df_cleaned['x45'] + df_cleaned['x46'] + df_cleaned['x47']
)

df_cleaned['avg_renting_cost'] = (
    df_cleaned['x48'] + df_cleaned['x49'] + df_cleaned['x50'] + df_cleaned['x51']
) / 4

df_cleaned['avg_housing_purchase_cost'] = (
    df_cleaned['x52'] + df_cleaned['x53'] + df_cleaned['x55']
) / 3

df_cleaned ['avg_salary'] = (df_cleaned['x54'])
```

```python
columns_to_drop = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'x16', 'x17', 'x18', 'x19', 'x20', 
                   'x21', 'x22', 'x23', 'x24', 'x25', 'x26', 'x27', 'x28', 'x29', 'x30', 'x31', 'x32', 'x33', 'x34', 'x35', 'x36', 'x37', 'x38', 'x39', 'x40', 'x41', 'x42', 'x43', 'x44',
                   'x45', 'x46', 'x47', 'x48', 'x49', 'x50', 'x51', 'x52', 'x53', 'x53','x54', 'x55']
df_cleaned.drop(columns=columns_to_drop, inplace=True)
```

```python
df_cleaned['overall_cost'] = (
    df_cleaned['dining_out_cost'] + df_cleaned['grocery_cost'] + df_cleaned['avg_transport_cost'] + df_cleaned['utility_cost'] + df_cleaned['avg_entertainment_cost'] +
df_cleaned['education_cost'] + df_cleaned['clothing_cost'] + df_cleaned['avg_renting_cost'] + df_cleaned['avg_housing_purchase_cost'])
)
```

```python
df_cleaned.columns
```

Handling outliers:

```python
df_cleaned = df_cleaned[(np.abs(stats.zscore(df_cleaned.select_dtypes(include=[np.number]))) < 3).all(axis=1)]
```

Normalizing the data:

```python
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_cleaned.select_dtypes(include=[np.number])), 
                         columns=df_cleaned.select_dtypes(include=[np.number]).columns)
```

Saving the new cleaned dataset:

```python
df_cleaned.to_csv('cleaned_cost_of_living_dataset.csv', index=False)
```
